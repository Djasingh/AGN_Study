```python
def researcher_dism_thesis(idd, ment):
    tmp = ment[ment['resId']==idd].copy()
    if tmp.shape[0]==1:
        return 'one'
    else:
        date_uniq = pd.unique(tmp['date_submitted'])
        uniq_thesisid = pd.unique(tmp['N_thesisId'])
        uniq_resid = pd.unique(tmp['resId'])
        if uniq_resid.shape[0]==1 and uniq_thesisid.shape[0]==1:
            return 'two'
        elif uniq_resid.shape[0]==1 and date_uniq.shape[0]==1: # uniq_thesisid.shape[0] > 1
            ment.loc[(ment['N_thesisId'].isin(uniq_thesisid)),'N_thesisId'] = uniq_thesisid[0]
            return 'three'     
        elif uniq_resid.shape[0]==1 and uniq_thesisid.shape[0] > 1 and date_uniq.shape[0] > 1 :
            if uniq_thesisid.shape[0] > date_uniq.shape[0]:
                for date in date_uniq:
                    if type(date)==str:
                        tid1=tmp[tmp['date_submitted']==date]['N_thesisId'].values[0]
                        tmp.loc[tmp['date_submitted']==date,'N_thesisId']=tid1
                        ment.loc[((ment['resId']==idd) & (ment['date_submitted']==date)),'N_thesisId']=tid1
                    else:
                        continue
            as_adv = ment[ment['advId']==idd].copy()
            uniq_stud=tmp.groupby(['N_thesisId'])['title'].apply(list).reset_index().copy()
            uniq_stud['title']=uniq_stud['title'].apply(lambda x: x[0])
            #print(uniq_stud['N_thesisId'])
            #as_stud_uniq_thesis = tmp[tmp['N_thesisId'].isin(uniq_thesisid)][['title','N_thesisId']]
            as_adv_uniq_thesis = pd.unique(as_adv['title'])
            if as_adv_uniq_thesis.shape[0] > 0:
                #print('yes')
                score=[]
                for index, stud in uniq_stud.iterrows():
                    score_list = [nlp(stud['title']).similarity(nlp(adv_thesis)) for adv_thesis in as_adv_uniq_thesis]
                    score.append((stud['N_thesisId'], sum(score_list)))
                sorted_score = sorted(score, key=lambda x: x[1],reverse=True)
                #print(sorted_score)
                #max_stud = tmp[tmp['title']==sorted_score[0][0]]
                for index1, stud1 in uniq_stud.iterrows(): 
                    if stud1['N_thesisId'] != sorted_score[0][0]:
                        ment.loc[((ment['resId']==idd) & (ment['N_thesisId']==stud1['N_thesisId'])), 'resId']=max(ment["resId"].max(axis=0), ment["advId"].max(axis=0))+1
                    else:    
                        continue
            else:
                count=1
                uniq_thesisid1=pd.unique(tmp['N_thesisId'])
                #print(uniq_thesisid1)
                for tid in uniq_thesisid1:
                    if count > 1 :
                        ment.loc[((ment['resId']==idd) & (ment['N_thesisId']==tid)),'resId']= max(ment["resId"].max(axis=0), ment["advId"].max(axis=0))+1
                        count += 1
                    else:
                        count += 1
                        continue
        else:
            print("Not Possible")
    return
```
