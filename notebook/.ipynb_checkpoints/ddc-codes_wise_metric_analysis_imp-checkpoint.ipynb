<h2> Fecundity Distrubution over DDC Codes </h2>

```python
def fecundity_comp(graph,df):
    fecundity=graph.out_degree()
    fecundity=dict(fecundity)
    inst_fecund_metric=inst_wise_val(df,'upper_ddc_code',fecundity)
    top_10, top_10_dict=top_metric_val(inst_fecund_metric)
    return inst_fecund_metric,top_10_dict
```

```python
ddc_fecundity1,ddc_top_fecundity1=fecundity_comp(graph1, filter_data1)
ddc_fecundity1= dict(sorted(ddc_fecundity1.items()))
```

```python
ddc_fecundity2,ddc_top_fecundity2=fecundity_comp(graph2, filter_data2)
ddc_fecundity2= dict(sorted(ddc_fecundity2.items()))
```

```python
ddc_fecundity3,ddc_top_fecundity3=fecundity_comp(graph3, filter_data3)
ddc_fecundity3= dict(sorted(ddc_fecundity3.items()))
```

```python
ddc_fecundity4,ddc_top_fecundity4=fecundity_comp(graph4, mod_ment2)
ddc_fecundity4= dict(sorted(ddc_fecundity4.items()))
```

<h2> Fertility Distrubution over DDC Codes </h2>

```python
def fertility_comp(graph,df):
    fertility=fertility_calculation(graph)
    fertility=dict(fertility)
    inst_fertility_metric=inst_wise_val(df,'upper_ddc_code',fertility)
    top_10, top_10_dict=top_metric_val(inst_fertility_metric)
    return inst_fertility_metric,top_10_dict
```

```python
ddc_fertility1,_ = fertility_comp(graph1, filter_data1)
ddc_fertility1= dict(sorted(ddc_fertility1.items()))
```

```python
ddc_fertility2,_=fertility_comp(graph2, filter_data2)
ddc_fertility2= dict(sorted(ddc_fertility2.items()))
```

```python
ddc_fertility3, _=fertility_comp(graph3, filter_data3)
ddc_fertility3= dict(sorted(ddc_fertility3.items()))
```

```python
ddc_fertility4, _=fertility_comp(graph4, mod_ment2)
ddc_fertility4= dict(sorted(ddc_fertility4.items()))
```
