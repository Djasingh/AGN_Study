```python
import re
import pickle
import pandas as pd
from tqdm import tqdm
```

```python
def load_obj(name):
    with open(name, 'rb') as f:
        return pickle.load(f)
```

```python
def obtain_uniq_ddc(code_list):
    ddc_code=[a.split(";") if ";" in a else [a] for a in code_list]
    ddc_code_lst= [item for sublist in ddc_code for item in sublist]
    return set(ddc_code_lst)
```

```python
def find_number(text):
    text=str(text)
    num = re.findall(r'\d+',text)
    return ";".join(num)
```

```python
sim_dept=load_obj('disambigauted_dept_wid_3991725.pkl')
```

```python
len(sim_dept)
```

```python
remove = [('department of mechanical engineering','department of chemical engineering'),('department of organic chemistry','department of inorganic chemistry'),('department of chemical engineering','department of mechanicalengineering'),('department of chemical oceanography','department of physical oceanography')]
```

```python
f_sim_names = [((a[0][0],a[1][0]),(sim_dept[a][0], sim_dept[a][1])) for a in sim_dept if (sim_dept[a][2] >= 10 or sim_dept[a][3] >= 10)]
```

```python
f_sim_names = [a for a in f_sim_names if a[1] not in remove]
```

```python
len(f_sim_names)
```

```python
ment['N_DepartmentId'] = ment['DepartmentId']
tmp_dict={}
for a in f_sim_names:
    if a[0][0] in tmp_dict and a[0][1] in tmp_dict:
        ment.loc[(ment['N_DepartmentId'].isin([tmp_dict[a[0][0]], tmp_dict[a[0][1]]])),'N_DepartmentId']= tmp_dict[a[0][0]]
    elif a[0][0] in tmp_dict:
        ment.loc[(ment['N_DepartmentId'].isin([tmp_dict[a[0][0]], a[0][1]])),'N_DepartmentId'] = tmp_dict[a[0][0]]
    elif a[0][1] in tmp_dict:
        ment.loc[(ment['N_DepartmentId'].isin([a[0][0], tmp_dict[a[0][1]]])),'N_DepartmentId'] = a[0][0]
    else:
        tmp_dict[a[0][1]]=a[0][0]
        ment.loc[(ment['N_DepartmentId'].isin([a[0][0], a[0][1]])),'N_DepartmentId'] = a[0][0]
```

```python
pd.unique(ment['N_DepartmentId']).shape
```

```python
ment['ddc_subject_code'] = ment['dc.subject.ddc'].apply(lambda x: find_number(x))
```

```python
dept_ddc = ment.groupby(['instituteId','DepartmentId'])['ddc_subject_code'].apply(list).reset_index().copy()
```

```python
dept_ddc['uniq_ddc_across_dept'] = dept_ddc['ddc_subject_code'].apply(lambda x: obtain_uniq_ddc(x))
```

```python
dept_ddc['uniq_ddc_across_dept'] = dept_ddc['uniq_ddc_across_dept'].apply(lambda x: sorted(x))
```

```python
dept_ddc['uniq_ddc_across_dept'] = dept_ddc['uniq_ddc_across_dept'].apply(lambda x: tuple(x))
```

```python
pd.unique(dept_ddc['uniq_ddc_across_dept'])
```

```python
dept_ddc[(dept_ddc['instituteId']=='I251') & (dept_ddc['N_DepartmentId']=='D2490')]
```

```python
dept_ddc['N_DepartmentId_1']=dept_ddc['N_DepartmentId']
uniq_inst = pd.unique(dept_ddc['instituteId'])
for inst in tqdm(uniq_inst):
    inst_data = dept_ddc[dept_ddc['instituteId']==inst]
    if inst_data.shape[0] > 1:
        #print(inst_data)
        uniq_ddc=pd.unique(inst_data['uniq_ddc_across_dept'])
        for ddc in uniq_ddc:
            #print(ddc)
            if len(ddc) > 0:
                sim_deptid = inst_data[inst_data['uniq_ddc_across_dept']==ddc]['N_DepartmentId_1'].tolist()
                #print(sim_deptid)
                dept_ddc.loc[(dept_ddc['instituteId']==inst) & (dept_ddc['uniq_ddc_across_dept']==ddc),'N_DepartmentId_1']=sim_deptid[-1]
            else:
                continue
    else:
        continue
```

```python
ment1=ment.merge(dept_ddc, left_on=['instituteId', 'N_DepartmentId'], right_on=['instituteId', 'N_DepartmentId'])
```
