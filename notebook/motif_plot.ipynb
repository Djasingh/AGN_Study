```python
import matplotlib.pyplot as plt
```

```python
motifs_name = ['3-node', '4-node', '5-node']
motifs_ferq = [224, 28, 241]
```

```python
fig = plt.figure(figsize = (12, 8))
 
# creating the bar plot
bars= plt.bar(motifs_name, motifs_ferq, color ='b',width = 0.4)
plt.yticks(fontsize=18,fontweight='bold')
plt.xticks(fontsize=18,fontweight='bold')
plt.xlabel("Subgraphs", fontsize=20,fontweight='bold')
plt.ylabel("Frequency", fontsize=20,fontweight='bold')

for bar in bars:
    plt.text(bar.get_xy()[0]+(bar.get_width()-0.3),bar.get_height() , str(bar.get_height()), color='green', fontweight='bold',fontsize=16)
plt.title("")
plt.savefig('../graphs_v5_2_7_2/motifs_v5_2_7_2.pdf',bbox_inches='tight')
plt.show()
```
